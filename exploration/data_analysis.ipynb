```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def load_jsonl_to_df(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            data.append(json.loads(line.strip()))
    return pd.DataFrame(data)

file_path = "/home/onyxia/work/SAVs/data/naturalbench_ret_train.jsonl"
df = load_jsonl_to_df(file_path)

df.head()
```

```python
def plot_label_counts(df):
    label_counts = Counter(df['label'])
    
    plt.figure(figsize=(8,6))
    plt.bar(label_counts.keys(), label_counts.values(), color='skyblue')
    plt.xlabel('Labels')
    plt.ylabel('Count')
    plt.title('Label Distribution')
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_label_counts(df)
```

```python
def plot_question_length_distribution(df):
    question_lengths = df['question'].apply(lambda x: len(x.split()))
    plt.figure(figsize=(8,6))
    plt.hist(question_lengths, bins=20, color='lightgreen', edgecolor='black')
    plt.xlabel("Length of Question (in words)")
    plt.ylabel("Frequency")
    plt.title("Distribution of Question Lengths")
    plt.show()
```

```python
plot_question_length_distribution(df)
```
