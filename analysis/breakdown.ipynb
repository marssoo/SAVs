```python
import pandas as pd
```

```python
df = pd.read_csv('result_csvs/all_val.csv')
```

```python
acc_columns = df.filter(like='_acc').drop(columns=['pets_acc', 'eurosat_acc'])
df['avg_acc'] = acc_columns.mean(axis=1)
```

```python
df.sort_values('avg_acc', ascending=False).head(10)
```

```python
for col in df.columns:
    if 'acc' not in col:
        continue
    df_restricted = df[['model', 'score', 'hp1', 'hp2', col]]
    print(f'############### on {col} ###############')
    print(df_restricted.sort_values(col, ascending=False).head(5))
```
