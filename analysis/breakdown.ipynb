```python
import pandas as pd
```

```python
#getting rid of the stds for now
df_val = pd.read_csv('result_csvs/all_val.csv').sort_values('avg_acc', ascending=False)
df_test = pd.read_csv('result_csvs/all_test.csv').sort_values('avg_acc', ascending=False)

df_val = df_val.loc[:, ~df_val.columns.str.contains('_std')]
df_test = df_test.loc[:, ~df_test.columns.str.contains('_std')]
```

```python
blink_columns = df_test.filter(like='blink')
blink_avg = blink_columns.mean(axis=1)
df_test = df_test.loc[:, ~df_test.columns.str.contains('blink')]
df_test['blink_avg'] = blink_avg
```

```python
new_order = [
    'model', 'num_heads', 'score', 'hp1', 'hp2',
    'MHalu_acc', 'vlguard_acc', 'blink_avg',
    'natural_ret_raw_acc', 'natural_ret_q_acc', 'natural_ret_i_acc', 'natural_ret_g_acc',
    'eurosat_acc', 'pets_acc'
]

df_test = df_test[new_order].drop(columns=['natural_ret_raw_acc'])
```
